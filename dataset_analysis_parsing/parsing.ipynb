```python
import pandas as pd
import string
import re
import datetime
```

```python
def parse(text):
    no_punctuation = text.translate(str.maketrans('', '', string.punctuation))
    no_small_words = re.sub(r'\b\w{1,3}\b', '', no_punctuation.lower())
    no_big_spaces = re.sub(' +', ' ', no_small_words)
    return no_big_spaces.strip()
```

```python
for i in range(0, len(df)):
    df.at[i, 'Text'] = parse(df.at[i, 'Text'])
    try:
        date_time = datetime.datetime.fromtimestamp(int(df.at[i, 'Time']))
    except ValueError:
        continue
    date = date_time.strftime('%Y-%m-%d %H:%M:%S')
    year = date.split('-')[0]
    try:
        int_year = int(year)
    except ValueError:
        continue
    df.at[i, 'Time'] = int_year
```

```python
diz = {}
for i in range(1, len(df)):
    if (df.at[i, 'Time'], df.at[i, 'ProductId']) not in diz:
        diz[(df.at[i, 'Time'], df.at[i, 'ProductId'])] = df.at[i, 'Text']
    else:
        diz[(df.at[i, 'Time'], df.at[i, 'ProductId'])] = diz[(df.at[i, 'Time'], df.at[i, 'ProductId'])] + ' ' + str(df.at[i, 'Text'])
```

```python
year_product2wordcount = {}
for (year, product) in diz:
    word2count = {}
    words = diz[year, product].split(' ')
    for word in words:
        if word not in word2count:
            word2count[word] = 0
        word2count[word] += 1
    
    year_product2wordcount[year, product] = word2count
```

```python
sorted_final = {}
for i in year_product2wordcount:
    sorted_final[i] = {k: v for k, v in sorted(year_product2wordcount[i].items(),
                                                                 key=lambda item: item[1], reverse=True)}
```

```python
finale = {}
```

```python
for i in sorted_final:
    finale[i] = (list((sorted_final[i].keys()))[:5], list(map(str, list((sorted_final[i].values()))))[:5])
```
