```python
import pandas as pd
import string
import re
import datetime
```

```python
def parse(text):
    no_punctuation = text.translate(str.maketrans('', '', string.punctuation))
    no_small_words = re.sub(r'\b\w{1,3}\b', '', no_punctuation.lower())
    no_big_spaces = re.sub(' +', ' ', no_small_words)
    return no_big_spaces
```

```python
df = df.drop(['ProfileName', 'Summary'], axis = 1)
```

```python
for i in range(0, len(df)):
    df.at[i, 'Text'] = parse(df.at[i, 'Text'])
    try:
        date_time = datetime.datetime.fromtimestamp(int(df.at[i, 'Time']))
    except ValueError:
        continue
    date = date_time.strftime('%Y-%m-%d %H:%M:%S')
    year = date.split('-')[0]
    try:
        int_year = int(year)
    except ValueError:
        continue
    df.at[i, 'Time'] = int_year
```
